#### Given an array, count the occurence of each unique element

```python
a= np.random.randint(1,10, size=(5,5))
a
```

```python
Values= np.unique(a, return_counts= True)[0]
Counts= np.unique(a, return_counts= True)[1]
```

```python
for i in range(len(Values)):
    print(Values[i],Counts[i],end="")
    print()
```

#### You have a 2D NumPy array where each row represents a customer's purchase history. The first column is the customer ID, and the second column is the total amount spent on purchases. You need to: purchase_data = np.array([[1, 1200], [2, 850], [3, 1500], [4, 600], [5, 2000], [6, 950]]) Identify customers who have spent more than $1000 in total. Group customers into two categories: "High Spend" (greater than $1000) and "Low Spend" (less than or equal to $1000)


```python
purchase_data = np.array([[1, 1200], [2, 850], [3, 1500], [4, 600], [5, 2000], [6, 950]]) 
purchase_data
```

```python
# spend more than 1000
print(f"Customer id of more than 1000 : {purchase_data[purchase_data[:,1]>1000][:,0]}")

# grouping
b= np.where(purchase_data[:,1]>1000, "High Spend" , "Low Spend")
```

```python
np.concatenate((purchase_data[:,0].reshape(6,1),b.reshape(6,1)), axis=1)
```

```python
purchase_data[:,0].reshape(6,1)
```
