# Pandas Series

A Series is a **one-dimensional** labeled array capable of holding **any data type**. The axis labels are collectively called the **index**.

**Creating Series**

**Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() # basically returns True where value is null otherwise False
```

```python
df.isna().sum() # find the total number of null present in every column
```

```python
df.isna().any()
```

**Removing Missing Data**

```python
df
```

```python
df.dropna() #removing missing values works along row 
```

```python
df.dropna(thresh=3) 
```

thresh basically have a numeric value.. the numeric significes the minimum number of non null values then the row will not be dropped

 Above we see 0,1,4 index rows have atleast 3 non null values thats why it is showing 
 2,3 is dropped because they have only 2 non null

**Filling the missing Data**

```python
df
```

```python
df.fillna(0) # fillna - fills the null values with a value, works along column
```

```python
values = {'A':0,'B':100,"C":300,'D':400} # Jis column name m jo null value h wo iss se replace hojayega
df.fillna(value=values)
```

```python
df.fillna(df.mean()) 
```

***isna - find missing, dropna - drop missing, fillna - fill missing***

***isna - along column, dropna - along row, fillna - along column***

**Merging 2 dataframes**

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how ='inner') #by default how value is inner
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right')
```

**GroupBy**

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by Category and calculate the sum of Sales
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group by Store and calculate the sum of Sales
cat = df.groupby('Store')['Sales'].sum()
cat
```

```python
# Group by multiple columns
# Group by Category and Store
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```
